```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Disable the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
zomato_df= pd.read_csv("E:\Python\zomato_restaurants.csv")
zomato_df
```

### a) Plot the Count of Famous Food Outlets / Restaurants in Bengaluru

```python
plt.figure(figsize=(10, 7))
sns.set_style('darkgrid')
famous= zomato_df['name'].value_counts()[:10]
sns.barplot(x= famous, y = famous.index, palette= 'dark')
sns.despine()
plt.title('Famous Food Outlets / Restaurants in Bengaluru')
plt.xlabel('Number of outlets')
plt.ylabel('Name of Restaurants')
```

### b) Display accepting online orders vs not accepting online orders count % distribution

```python
plt.figure(figsize= (5,5))
plt.pie(zomato_df['online_order'].value_counts(),labels=zomato_df['online_order'].unique(),autopct='%1.2f%%',explode = [0,0])
plt.legend()
plt.show()
```

### d) Display accepting Table Booking vs Not accepting table booking count % distribution

```python
plt.figure(figsize= (5,5))
plt.pie(zomato_df['book_table'].value_counts(),labels=zomato_df['book_table'].unique(),autopct='%1.2f%%',explode = [0,0])
plt.legend()
plt.show()
```

### c) Find the popular restaurant type in Bengaluru.

```python
rest_type= zomato_df['rest_type'].value_counts()[:20]
plt.figure(figsize=(8, 6))
sns.set_style('darkgrid')
sns.barplot(x= rest_type, y = rest_type.index)
plt.title('popular restaurant type in Bengaluru')
plt.xlabel('Count')
plt.ylabel('Restaurant type')
```

### e) Histogram and Box Plot of Approx Cost Distribution. Note: Replace â€˜,â€™ from the values using replace(), lambda function with replace()


```python
cost= zomato_df['approx_cost(for two people)']
cost= cost.dropna().apply(lambda x : int(x.replace(',', '')))
plt.figure(figsize=(6,6))
sns.set_style('darkgrid')
sns.distplot(cost, color= 'green')
plt.title('Approx. cost for two people')
plt.xlabel('Cost for two people')
```

### f) Histogram and Box Plot of Restaurant Rating Distribution. Note: Replace â€˜,â€™ from the values using replace(), lambda function with replace()

```python
rating= zomato_df['rate']
rating= rating.dropna().apply(lambda x : float(x.split('/')[0]) if (len(x)>3) else np.nan).dropna()
plt.figure(figsize=(7,6))
sns.set_style('darkgrid')
sns.distplot(rating, bins = 20,  color= 'red')
plt.title('Rating distribution of the restaurants')
plt.xlabel('Ratings')
```

### g) Rate vs Approx Cost (for two people). Also categorise them into online_order available or not. 

```python
crdf= zomato_df[['rate', 'approx_cost(for two people)', 'online_order']].dropna()
crdf['rate']= crdf['rate'].apply(lambda x : float(x.split('/')[0]) if (len(x)>3) else 0)
crdf['approx_cost(for two people)']= crdf['approx_cost(for two people)'].apply(lambda x : int(x.replace(',', '')))

plt.figure(figsize=(11, 8))
sns.set_style('darkgrid')
sns.scatterplot( x= 'rate', y = 'approx_cost(for two people)', hue= 'online_order', data= crdf )
plt.title('Cost vs Rating comparison')
plt.xlabel('Rating between 0 to 5')
plt.ylabel('Approx. cost for 2 people')
```
