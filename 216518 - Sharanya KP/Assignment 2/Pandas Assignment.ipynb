### 2. How to stack two random series vertically and horizontally?

```python
series1 = pd.Series(['g', 'e', 'e', 'k', 's'])
series2 = pd.Series([9, 8, 7, 6, 5])
   
# stacking the series horizontally
df = pd.concat([series1, series2], axis = 1)
print("Stack two series horizontally:")
display(df)
```

```python
# stacking the series vertically
df = pd.concat([series1, series2], axis = 0)
print("Stack two series vertically:")
display(df)
```

### 6. How to filter valid emails from a series data using following pattern to be matched?
    pattern ='[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'

	create series of email id by:
      emails = pd.Series(['books at amazom.com', 'james@egypt.com', 'jones@yahoo.in', 'bigdata@gmail.com'])


```python
emails = pd.Series(['books at amazom.com', 'james@egypt.com', 'jones@yahoo.in', 'bigdata@gmail.com'])

#initializing valid email pattern
pattern ='[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\\.[A-Za-z]{2,4}'

#finding all matching occurrences
result = emails.str.findall(pattern)

print("Valid Emails are: ")
print([i for i in result if len(i) > 0])
```
